```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

data = pd.read_csv('trainRegression.csv') 

data.head()

print (data['X'])

#plt.plot(data['X'],data['R'],'*')

type(data)

# to convert data into Numpy Array
#numpydata = data.as_matrix()
#type(numpydata)

# data = np.array(data, dtype=np.float)

numpy_array = data.to_numpy()

print(type(numpy_array))

```

```python
test_data = pd.read_csv('testRegression.csv') 

test_data.head()

print(test_data['X'])

#plt.plot(test_data['X'],test_data['R'],'*')

type(test_data)

numpy_array = test_data.to_numpy()

print(type(numpy_array))
```

```python
#for linear regression model(training data)
x=np.array(data['X'])
y=np.array(data['R'])


#total no of sample
m=np.size(x)

#sum of x
xi=sum(x)


#sum of x
xi=sum(x)


#multiply of x and x
xx=sum(x*x)


#sum of y
yi=sum(y)


#multiply of x and y
yixi=(np.dot(x,y))


mat_A = np.array([[m, xi], [xi, xx]])
mat_B = np.array([[yi], [yixi]])


print("MATRIX A=\n",mat_A)
print("\nMATRIX B=\n",mat_B)




mat_X=np.dot((np.linalg.inv(mat_A)),mat_B)
print("\nMATRIX X=\n",mat_X) 


#hypothesis function for training
yprime=mat_X[0][0]+mat_X[1][0]*x
print("\n",yprime)
#graph for training data
#plt.plot(x,y,'+')
#plt.plot(x,yprime,'*')

#formula for mean square error
msqe_tr=(sum(np.square(yprime-y)))/(2*m)
print("\nLinear Mean square error for training data:",msqe_tr)

#print("\nGraph for training data of linear regression")


```

```python
#for linear regression model(testing data)
x_one=np.array(test_data['X'])
y_one=np.array(test_data['R'])

#total no of sample
m_one=np.size(x_one)

#hypothesis function for testing
yprime_one=mat_X[0][0]+mat_X[1][0]*x_one

print(yprime_one)

#graph for training data
plt.plot(x_one,y_one,'*')
plt.plot(x_one,yprime_one,'+')

#formula for mean square error (testing data)
msqe_tr=(sum(np.square(yprime_one-y_one)))/(2*m_one)
print("\nLinear Mean square error for testing data:",msqe_tr)


print("\nGraph for testing data of linear regression")
```

```python
#for quadratic regression model(training data)
x=np.array(data['X'])
y=np.array(data['R'])

#total no of sample
m=np.size(x)

#xikacube
xikacube=sum(x*x*x)

#xikafour
xikafour=sum(x*x*x*x)


#multiply of yi and x2
xxyi=sum(x*x*y)

mat_A = np.array([[m, xi, xx], [xi, xx, xikacube], [xx, xikacube, xikafour]])
mat_B = np.array([[yi], [yixi], [xxyi]])


print("MATRIX A=\n",mat_A)
print("\nMATRIX B=\n",mat_B)


mat_X=np.dot((np.linalg.inv(mat_A)),mat_B)
print("\nMATRIX X=\n",mat_X) 

#hypothesis function for training
yprime=mat_X[0][0]+mat_X[1][0]*x+mat_X[2][0]*(x*x)
print("\n",yprime)

#graph for training data
#plt.plot(x,y,'+')
#plt.plot(x,yprime,'*')

#formula for mean square error
msqe_tr=(sum(np.square(yprime-y)))/(2*m)
print("\nQuadratic Mean square error for training data:",msqe_tr)

#print("\nGraph for training data of quadratic regression")
```

```python
#for quadratic regression model(testing data)
x_one=np.array(test_data['X'])
y_one=np.array(test_data['R'])

#hypothesis function for testing
yprime_one=mat_X[0][0]+mat_X[1][0]*x_one+mat_X[2][0]*(x_one*x_one)

print(yprime_one)

#graph for training data
plt.plot(x_one,y_one,'*')
plt.plot(x_one,yprime_one,'+')

#formula for mean square error (testing data)
msqe_tr=(sum(np.square(yprime_one-y_one)))/(2*m_one)
print("\nQuadratic Mean square error for testing data:",msqe_tr)


print("\nGraph for testing data of quadratic regression")
```

```python
#for cubic regression model(training data)
x=np.array(data['X'])
y=np.array(data['R'])

#total no of sample
m=np.size(x)

#xikafive
xikafive=sum(x*x*x*x*x)

#xikasix
xikasix=sum(x*x*x*x*x*x)

#multiply of yi and x3
xxxyi=sum(x*x*x*y)

mat_A = np.array([[m, xi, xx, xikacube], [xi, xx, xikacube, xikafour], [xx, xikacube, xikafour, xikafive], [xikacube, xikafour, xikafive, xikasix]])
mat_B = np.array([[yi], [yixi], [xxyi], [xxxyi]])


print("\nMATRIX A=\n",mat_A)
print("\nMATRIX B=\n",mat_B)


mat_X=np.dot((np.linalg.inv(mat_A)),mat_B)
print("\nMATRIX X=\n",mat_X) 

#hypothesis function for training
yprime=mat_X[0][0]+mat_X[1][0]*x+mat_X[2][0]*(x*x)+mat_X[3][0]*(x*x*x)
print("\n",yprime)

#graph for training data
#plt.plot(x,y,'+')
#plt.plot(x,yprime,'*')

#formula for mean square error
msqe_tr=(sum(np.square(yprime-y)))/(2*m)
print("\nCubic Mean square error for training data:",msqe_tr)

#print("\nGraph for training data of cubic regression")
```

```python
#for cubic regression model(testing data)
x_one=np.array(test_data['X'])
y_one=np.array(test_data['R'])

#hypothesis function for testing
yprime_one=mat_X[0][0]+mat_X[1][0]*(x_one)+mat_X[2][0]*(x_one*x_one)+mat_X[3][0]*(x_one*x_one*x_one)

print(yprime_one)

#graph for training data
plt.plot(x_one,y_one,'*')
plt.plot(x_one,yprime_one,'+')

#formula for mean square error (testing data)
msqe_tr=(sum(np.square(yprime_one-y_one)))/(2*m_one)
print("\nCubic Mean square error for testing data:",msqe_tr)


print("\nGraph for testing data of cubic regression")
```
